1.	 <b>Pengertian Machine Learning</b> <br>
Machine learning adalah cabang dari AI (artificial Intelegence) atau kecerdsan buatan yang lebih berfokus pada pengembangan sistem yang dapat belajar dari banyak data. Dari pembelajaran itulah yang akan dijadikan sebuah pengambilan keputusan. Semakin luas hal yang dipelajari maka makin akurat pengambilan keputusan yang diberikan oleh mesin. Tujuan Machine Learning adalah kemampuan untuk memproses data dalam jumlah besar dengan cepat dan akurat, sehingga menghasilkan prediksi, keputusan dan hasil yang baik. Penerapan Machine Learning memmungkinkan sistem untuk belajar dari pengalaman tanpa perlu pemrograman manual yang kompleks. Sehingga dapat beradaptasi dengan perubahan dan memperbaiki kinerja.

3.	 <b>Taxonomi dalam pengerapan Machine Learning</b> <br>
•	Supervised learning
Supervised learning adalah jenis machine learning yang menggunakan data yang telah terlabel dan memprediksi output yang sesuai dengan input yang diberikan. Contoh aplikasi supervised learning adalah sistem rekomendasi yang memprediksi film atau serial TV yang mungkin disukai pengguna berdasarkan data yang telah dikumpulkan tentang film yang telah ditonton.
Supervised learning ini terbagi menjadi beberapa jenis:
-	Regresi <br>
Untuk mempelajari fungsi yang dapat menghubungkan fitur input dengan nilai kontinu yang terkait, seperti memprediksi harga rumah berdasarkan ukuran, lokasi, dan fitur lainnya.
-	Klasifikasi<br>
Untuk mempelajari fungsi yang dapat menghubungkan fitur input dengan kelas output yang terkait, seperti memprediksi apakah email masuk ke dalam kotak masuk atau kotak spam.
-	Deteksi Anomali<br>
Untuk mempelajari pola data normal dan menggunakan pola tersebut untuk mendeteksi data yang tidak biasa atau aneh.
-	Segmentasi<br>
Untuk mempelajari pola-pola pada data dan mengelompokkan data ke dalam segmen atau kelompok yang berbeda.
-	Prediksi Rangkaian Waktu<br>
Untuk mempelajari pola-pola pada data bersejarah dan menggunakan pola tersebut untuk memprediksi nilai-nilai pada waktu yang akan datang.
-	Rekomendasi<br>
Untuk mempelajari preferensi pengguna dan menggunakan informasi ini untuk merekomendasikan produk atau item yang relevan.
-	Pengenalan Suara<br>
Untuk mempelajari pola suara dan menggunakan pola tersebut untuk mengenali kata-kata yang diucapkan.
-	Pengenalan Gambar<br>
Untuk mempelajari pola gambar dan menggunakan pola tersebut untuk mengenali objek, wajah, atau situasi yang terkait dengan gambar tersebut.

•	Unsupervised learning <br>
Unsupervised learning adalah jenis machine learning yang menggunakan data yang tidak terlabel dan mencari pola atau tren yang mungkin terjadi. Contoh aplikasi unsupervised learning adalah sistem yang dapat mengelompokkan produk berdasarkan karakteristik yang sama tanpa adanya instruksi secara eksplisit.
Unsupervised learning ini terbagi menjadi beberapa jenis:
-	Clustering, jenis unsupervised learning di mana data dipartisi menjadi kelompok atau klaster yang homogen berdasarkan kemiripan fitur.
-	Dimensionality Reduction, jenis unsupervised learning yang bertujuan untuk mengurangi dimensi data dengan memilih fitur-fitur yang paling signifikan atau mewakili data dengan cara yang lebih efisien.
•	Reinforcement learning<br>
Reinforcement learning adalah jenis machine learning yang mengajarkan sistem untuk mengambil tindakan yang tepat sesuai dengan situasi yang terjadi dengan memberikan reward atau punishment. Contoh aplikasi reinforcement learning adalah mobil tanpa pengemudi (self-driving car) yang dapat mempelajari cara mengambil keputusan yang tepat sesuai dengan situasi yang terjadi.
Macam Reinforcement learning:
-	Model-Based Reinforcement Learning
-	Model-Free Reinforcement Learning
-	Multi-Agent Reinforcement Learning

•	Transfer learning<br>
Transfer learning adalah jenis machine learning yang memanfaatkan pengetahuan yang telah dipelajari oleh sistem dari suatu domain untuk memprediksi keputusan dalam domain yang lain. Contoh aplikasi transfer learning adalah sistem yang dapat memprediksi keputusan pembelian produk berdasarkan data yang telah dikumpulkan tentang pembelian produk di toko lain.

Transfer Learning dibagi menjadi beberap jenis :
-	Transfer learning berdasarkan arsitektur model
-	Transfer learning berdasarkan sumber data
-	Transfer learning berdasarkan tugas
-	Transfer learning berdasarkan layer
•	Deep learning
Deep learning adalah jenis machine learning yang menggunakan jaringan saraf yang kompleks untuk memprediksi keputusan yang mungkin diambil. Contoh aplikasi deep learning adalah sistem yang dapat memprediksi penyakit yang mungkin terjadi pada individu dan memberikan saran terapi yang tepat.
